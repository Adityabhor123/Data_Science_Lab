### Factorial Calculation (Iterative Approach)

This code snippet defines a function to calculate the factorial of a non-negative integer using an iterative method. It also includes error handling for negative numbers and correctly handles the factorial of 0.

```python
def factorial_iterative(n):
    """Calculates the factorial of a non-negative integer using an iterative approach.

    Args:
        n (int): A non-negative integer.

    Returns:
        int: The factorial of n.

    Raises:
        ValueError: If n is a negative number.
    """
    if n < 0:
        raise ValueError('Factorial is not defined for negative numbers.')
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
print(f"Factorial of {number}: {factorial_iterative(number)}")
print(f"Factorial of {number_zero}: {factorial_iterative(number_zero)}")

try:
    print(f"Factorial of {number_negative}: {factorial_iterative(number_negative)}")
except ValueError as e:
    print(f"Factorial of {number_negative}: Error - {e}")
```

### Multiplication Table

This code snippet generates and displays the multiplication table for a number provided by the user, from 1 to 10.

```python
# Function to display multiplication table
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11): # Table from 1 to 10
        print(f"{num} x {i} = {num * i}")

# Get input from the user
try:
    number_to_table = int(input("Enter a number to display its multiplication table: "))
    display_multiplication_table(number_to_table)
except ValueError:
    print("Invalid input. Please enter an integer.")
```

### Odd or Even Number Identifier

This code determines whether a user-inputted integer is odd or even using the modulo operator (`%`).

```python
# Function to check if a number is odd or even
def check_odd_even(num):
    if num % 2 == 0:
        print(f"The number {num} is Even.")
    else:
        print(f"The number {num} is Odd.")

# Get input from the user
try:
    number_to_check = int(input("Enter an integer to check if it's odd or even: "))
    check_odd_even(number_to_check)
except ValueError:
    print("Invalid input. Please enter an integer.")
```

### Palindrome Number Identifier

This code checks if a user-inputted integer is a palindrome. A palindrome is a number that reads the same forwards and backwards.

```python
# Function to check if a number is a palindrome
def is_palindrome(num):
    # Convert the number to a string to easily reverse it
    num_str = str(num)
    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        print(f"The number {num} is a Palindrome.")
    else:
        print(f"The number {num} is not a Palindrome.")

# Get input from the user
try:
    number_to_check = int(input("Enter an integer to check if it's a palindrome: "))
    is_palindrome(number_to_check)
except ValueError:
    print("Invalid input. Please enter an integer.")
```

### Knowledge Base: Student Marks

To establish the knowledge base for student marks, we'll define two key components:

1.  **Grading Rules**: A list of dictionaries will represent the grading scale. Each dictionary will specify the `min_score`, `max_score`, `letter_grade`, and `performance` description for a particular grade range.

**Reasoning**:

```python
GRADING_RULES = [
    {"min_score": 90, "max_score": 100, "letter_grade": "A", "performance": "Excellent"},
    {"min_score": 80, "max_score": 89, "letter_grade": "B", "performance": "Very Good"},
    {"min_score": 70, "max_score": 79, "letter_grade": "C", "performance": "Good"},
    {"min_score": 60, "max_score": 69, "letter_grade": "D", "performance": "Pass"},
    {"min_score": 0, "max_score": 59, "letter_grade": "F", "performance": "Fail"}
]

MIN_PASS_SCORE = 60

print("Grading rules and minimum pass score defined.")
```

## Implement Inference Engine: Student Marks

### Subtask:
Write Python code to implement the inference engine for student marks. This component will take a student's score as input, consult the knowledge base (decision rules), and apply them to arrive at a conclusion or recommendation (e.g., 'A', 'Pass', 'Needs Improvement').


**Reasoning**:
I need to implement the inference engine for student marks by defining a Python function `infer_student_grade` that takes a score, validates it, applies the grading rules from `GRADING_RULES`, determines pass/fail status using `MIN_PASS_SCORE`, and returns a dictionary with the results.



```python
def infer_student_grade(score):
    """Infers the letter grade, performance description, and pass/fail status for a given student score.

    Args:
        score (int): The student's numerical score (0-100).

    Returns:
        dict: A dictionary containing 'letter_grade', 'performance', and 'pass_fail_status'.

    Raises:
        ValueError: If the score is not between 0 and 100.
    """
    # 2. Validate the input score
    if not (0 <= score <= 100):
        raise ValueError('Score must be between 0 and 100.')

    # 3. Initialize variables
    letter_grade = "N/A"
    performance = "N/A"
    pass_fail_status = "N/A"

    # 4. Iterate through the GRADING_RULES to determine letter_grade and performance
    for rule in GRADING_RULES:
        if rule["min_score"] <= score <= rule["max_score"]:
            letter_grade = rule["letter_grade"]
            performance = rule["performance"]
            break # Found the matching rule, no need to check further

    # 5. Determine the pass_fail_status
    if score >= MIN_PASS_SCORE:
        pass_fail_status = "Pass"
    else:
        pass_fail_status = "Fail"

    # 6. Return a dictionary containing the results
    return {
        "letter_grade": letter_grade,
        "performance": performance,
        "pass_fail_status": pass_fail_status
    }

print("Inference engine function 'infer_student_grade' defined.")
```
