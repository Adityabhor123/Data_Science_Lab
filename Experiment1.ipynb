```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

number_str = input("Enter a number to compute its factorial: ")
number = int(number_str)
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{num} x {i} = {num * i}")

number_to_display_str = input("Enter a number to display its multiplication table: ")
number_to_display = int(number_to_display_str)
display_multiplication_table(number_to_display)
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

num_input_str = input("Enter a number to check if it's odd or even: ")
num_input = int(num_input_str)
print(check_odd_even(num_input))
```

```python
def is_palindrome_number(number):
    # Convert the number to a string for easy reversal
    num_str = str(number)
    # Compare the original string with its reversed version
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome number."
    else:
        return f"{number} is not a palindrome number."

num_input_str = input("Enter a number to check if it's a palindrome: ")
num_input = int(num_input_str)
print(is_palindrome_number(num_input))
```

```python
def student_pass_fail_expert_system(marks):
    passing_marks = 50  # Define the passing threshold

    if marks >= passing_marks:
        return "Pass"
    else:
        return "Fail"

# Get student's marks from the user
marks_str = input("Enter student's marks: ")

try:
    marks = int(marks_str)
    if 0 <= marks <= 100: # Assuming marks are between 0 and 100
        result = student_pass_fail_expert_system(marks)
        print(f"With {marks} marks, the student's status is: {result}")
    else:
        print("Please enter marks between 0 and 100.")
except ValueError:
    print("Invalid input. Please enter a valid number for marks.")
```
