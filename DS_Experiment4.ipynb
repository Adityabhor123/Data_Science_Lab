```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a normal distribution with mean 0 and standard deviation 1
# We'll generate 1000 data points
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Add a title and labels
plt.title('Histogram of a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np

data = list(map(float,input("Enter the number to calculate the mean,mode,SD,variance").split()))

mean = np.mean(data)
mode = np.argmax(np.bincount(data))
SD = np.std(data)
variance = np.var(data)

print("mean:",mean)
print("mode:",mode)
print("SD:",SD)
print("variance:",variance)

```

```python
# Taking input from user
data = list(map(float, input("Enter numbers separated by space: ").split()))

n = len(data)

# Mean
mean = sum(data) / n

# Median
sorted_data = sorted(data)

if n % 2 == 0:
    median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
else:
    median = sorted_data[n//2]

# Population Variance
variance = sum((x - mean) ** 2 for x in data) / n

# Standard Deviation
std_dev = variance ** 0.5

print("\nManual Calculation:")
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
import random

# Take number of tosses from user
n = int(input("Enter number of coin tosses: "))

heads = 0
tails = 0

# Simulate coin tosses
for i in range(n):
    toss = random.choice(['H', 'T'])  # H = Heads, T = Tails

    if toss == 'H':
        heads += 1
    else:
        tails += 1

# Compute probabilities
prob_heads = heads / n
prob_tails = tails / n

# Display results
print("\n--- Results ---")
print("Total Tosses:", n)
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```
